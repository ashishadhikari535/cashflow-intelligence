```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import plotly.express as px
from IPython.display import display

from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

project_root = Path.cwd().resolve()
if project_root.name == "notebooks":
    project_root = project_root.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.utils.config import FEATURE_STORE_PATH, TEST_PATH, TRAIN_PATH
from src.models.classifier import get_Xy, load_model, predict
from src.models.evaluator import evaluate_classifier, risk_segment_analysis
from src.models.forecaster import (
    generate_forecast,
    plot_forecast,
    prepare_training_data,
    save_forecaster,
    train_forecaster,
    train_segment_forecasters,
)
from src.models.explainer import generate_risk_report

train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
X_train, y_train = get_Xy(train)
X_test, y_test = get_Xy(test)

print(f"Train rows: {len(train):,} | Test rows: {len(test):,}")
```

## 1) Classifier Baseline Comparison (`classifier.py` context)

Reasoning:
- Random baseline checks minimum signal.
- Logistic baseline checks linear separability.
- Trained XGBoost should outperform both.

```python
baselines = {}

dummy = DummyClassifier(strategy="stratified", random_state=42)
dummy.fit(X_train, y_train)
baselines["Random"] = roc_auc_score(y_test, dummy.predict_proba(X_test)[:, 1])

lr = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(max_iter=600, random_state=42)),
])
lr.fit(X_train, y_train)
baselines["Logistic"] = roc_auc_score(y_test, lr.predict_proba(X_test)[:, 1])

clf_model, clf_meta = load_model()
xgb_probs = clf_model.predict_proba(X_test)[:, 1]
baselines["XGBoost"] = roc_auc_score(y_test, xgb_probs)

fig = px.bar(
    x=list(baselines.keys()),
    y=list(baselines.values()),
    color=list(baselines.keys()),
    labels={"x": "Model", "y": "AUC-ROC"},
    title="Classifier Comparison on Held-out Test Set",
)
fig.add_hline(y=0.5, line_dash="dash", line_color="gray")
fig.update_layout(showlegend=False, yaxis_range=[0.4, 1.0])
fig.show()

for name, score in baselines.items():
    print(f"{name:<10} AUC-ROC: {score:.4f}")
```

## 2) Classifier Evaluation (`evaluator.py` parity)

This executes the same evaluation function used in the executable pipeline.

```python
metrics = evaluate_classifier(save_plots=True)

print("\n=== Evaluation Summary ===")
print(f"CV AUC-ROC: {metrics['cv_auc']:.4f}")
print(f"Test AUC-ROC: {metrics['auc_roc']:.4f}")
print(f"Test AUC-PR: {metrics['auc_pr']:.4f}")
print(f"Brier score: {metrics['brier']:.4f}")
print(f"Threshold: {metrics['threshold']:.2f}")
```

```python
seg_df = risk_segment_analysis()
display(seg_df)

print("\nReasoning: Segment-level AUC and flagged rate should remain stable across regions and industries.")
```

## 3) Business Impact View (classifier outputs)

Translate predictions into AR exposure terms.

```python
pred = predict(clf_model, test, clf_meta["threshold"])

total_ar = float(pred["DMBTR"].sum())
flagged = pred[pred["LATE_PRED"] == 1]
flagged_ar = float(flagged["DMBTR"].sum())
missed_late = pred[(pred["LATE_FLAG"] == 1) & (pred["LATE_PRED"] == 0)]
missed_ar = float(missed_late["DMBTR"].sum())

print("=" * 60)
print("BUSINESS IMPACT SUMMARY")
print("=" * 60)
print(f"Total AR in test set             : ${total_ar:,.0f}")
print(f"AR flagged for action            : ${flagged_ar:,.0f} ({flagged_ar / max(total_ar, 1):.1%})")
print(f"AR missed (late but not flagged) : ${missed_ar:,.0f} ({missed_ar / max(total_ar, 1):.1%})")
```

## 4) Forecast Pipeline (`forecaster.py` parity)

Notebook mode supports both:
- quick path: use saved forecaster artifact
- full path: retrain global + segment fallback models and save artifact

```python
RUN_FORECASTER_TRAIN = False

overall_df, pooled_df, segment_datasets, segment_summary = prepare_training_data()

summary_rows = []
for segment, stats in segment_summary.items():
    if segment == "__meta__":
        continue
    summary_rows.append({"segment": segment, **stats})

summary_df = pd.DataFrame(summary_rows).sort_values("segment").reset_index(drop=True)
display(summary_df.head(20))
print(f"Overall monthly rows: {len(overall_df)} | Pooled rows: {len(pooled_df)}")
print(f"Global training source: {segment_summary.get('__meta__', {}).get('global_training_source', 'unknown')}")

if RUN_FORECASTER_TRAIN:
    print("Training forecaster models...")
    global_models = train_forecaster(pooled_df)
    segment_models, segment_summary = train_segment_forecasters(segment_datasets, segment_summary)
    save_forecaster(
        global_models,
        overall_df,
        segment_models=segment_models,
        segment_summary=segment_summary,
    )
    print("Forecaster artifacts updated.")
else:
    print("Skipping retrain; using saved forecaster artifact.")
```

```python
forecast_df = generate_forecast(overall_df)
display(forecast_df)
plot_forecast(overall_df, forecast_df)
```

## 5) Explainability Pipeline (`explainer.py` parity)

Generate SHAP-backed narratives for top predicted late-risk invoices.

```python
TOP_N_RISKS = 10

feature_store = pd.read_csv(FEATURE_STORE_PATH, parse_dates=["BLDAT"])
risk_report = generate_risk_report(feature_store, top_n=TOP_N_RISKS)

if risk_report.empty:
    print("No invoices flagged at current threshold.")
else:
    cols = ["BELNR", "KUNNR", "LATE_PROB", "LATE_PRED", "RISK_LABEL", "NARRATIVE"]
    keep = [c for c in cols if c in risk_report.columns]
    display(risk_report[keep].head(5))
```
